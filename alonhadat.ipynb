```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://alonhadat.com.vn/nha-dat/can-ban/can-ho-chung-cu/2/ho-chi-minh/trang--"
session = requests.Session()

house_data_list = []

# L·∫∑p qua c√°c trang
for page in range(1, 50):  # B·∫°n c√≥ th·ªÉ thay ƒë·ªïi s·ªë trang t·∫°i ƒë√¢y
    response = session.get(url + str(page) + '.html')
    soup = BeautifulSoup(response.content, "html.parser")
    
    # T√¨m t·∫•t c·∫£ c√°c m·ª•c b·∫•t ƒë·ªông s·∫£n
    items = soup.find_all('div', class_='content-item item')
    for item in items:
        # L·∫•y URL c·ªßa trang chi ti·∫øt v√† ti√™u ƒë·ªÅ
        title_box = item.find('div', class_='ct_title_box title')
        title_tag = title_box.find('a') if title_box else None
        title = title_tag.get_text(strip=True) if title_tag else ""
        detail_url = title_tag['href'] if title_tag else ""
        print(detail_url)

        # Truy c·∫≠p v√†o URL chi ti·∫øt n·∫øu c√≥
        if detail_url:
            full_url = "https://alonhadat.com.vn" + detail_url
            detail_response = session.get(full_url)
            detail_soup = BeautifulSoup(detail_response.content, 'html.parser')
            
            # L·∫•y th√¥ng tin t·ª´ th·∫ª moreinfor
            more_info_tag = detail_soup.find('div', class_='moreinfor')
            gia = ""
            dientich = ""
            
            if more_info_tag:
                # L·∫•y gi√°
                gia_tag = more_info_tag.find('span', class_='price')
                gia = gia_tag.find('span', class_='value').get_text(strip=True) if gia_tag else ""
                
                # L·∫•y di·ªán t√≠ch
                dientich_tag = more_info_tag.find('span', class_='square')
                dientich = dientich_tag.find('span', class_='value').get_text(strip=True) if dientich_tag else ""
            
            # L·∫•y th√¥ng tin ƒë·ªãa ch·ªâ t·ª´ th·∫ª address
            address_tag = detail_soup.find('div', class_='address')
            address = address_tag.find('span', class_='value').get_text(strip=True) if address_tag else ""
            
            # L·∫•y c√°c th√¥ng tin kh√°c t·ª´ b·∫£ng infor
            infor_table = detail_soup.find('div', class_='infor')
            ma_tin = ""
            loai_tin = ""
            loai_bds = ""
            so_lau = ""
            so_phong_ngu = ""
            du_an = ""
            
            if infor_table:
                rows = infor_table.find_all('tr')
                for row in rows:
                    cols = row.find_all('td')
                    if len(cols) >= 2:
                        label = cols[0].get_text(strip=True)
                        value = cols[1].get_text(strip=True)
                        if label == "M√£ tin":
                            ma_tin = value
                        elif label == "Lo·∫°i tin":
                            loai_tin = value
                        elif label == "Lo·∫°i BDS":
                            loai_bds = value
                        elif label == "S·ªë l·∫ßu":
                            so_lau = value
                        elif label == "S·ªë ph√≤ng ng·ªß":
                            so_phong_ngu = value
                        elif label == "Thu·ªôc d·ª± √°n":
                            du_an = cols[1].find('a').get_text(strip=True) if cols[1].find('a') else ""
            
            # L∆∞u th√¥ng tin v√†o dictionary
            house_data = {
                'Ti√™u ƒë·ªÅ': title,
                'URL chi ti·∫øt': full_url,
                'Gi√°': gia,
                'Di·ªán t√≠ch': dientich,
                'ƒê·ªãa ch·ªâ': address,
                'M√£ tin': ma_tin,
                'Lo·∫°i tin': loai_tin,
                'Lo·∫°i BDS': loai_bds,
                'S·ªë l·∫ßu': so_lau,
                'S·ªë ph√≤ng ng·ªß': so_phong_ngu,
                'D·ª± √°n': du_an
            }
            
            # Th√™m d·ªØ li·ªáu v√†o danh s√°ch
            house_data_list.append(house_data)

df = pd.DataFrame(house_data_list)
print(df)
```
